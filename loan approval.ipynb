```python
#width
q1=df[' residential_assets_value'].quantile(0.25)
q3=df[' residential_assets_value'].quantile(0.75)
iqr=q3-q1
ub=q3+(1.5*iqr)
lb=q1-(1.5*iqr)
print("q1",q1)
print("q3",q3)
print("iqr",iqr)
print("upper_tail",ub)
print("lower_tail",lb)
```

```python
df.loc[df[' residential_assets_value']>ub , ' residential_assets_value']=ub
```

## Labelencoding


```python
obj_col = df.select_dtypes(include = "object")
obj_col.columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in obj_col:
    df[i] = le.fit_transform(df[i])
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
a = MinMaxScaler()
```
